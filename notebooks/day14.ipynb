```python
def make_map(s: str):
    rockmap = dict()
    min_x, max_x, min_y, max_y = None, None, None, None
    for line in s.splitlines():
        rock_formations = []
        for rock in line.split(" -> "):
            x,y = rock.split(",")
            rock_formations.append((int(x), int(y)))
            x = int(x)
            y = int(y)
            if min_x is None: min_x = x 
            elif min_x > x: min_x = x
            if max_x is None: max_x = x 
            elif max_x < x: max_x = x
            if min_y is None: min_y = y 
            elif min_y > y: min_y = y
            if max_y is None: max_y = y 
            elif max_y < y: max_y = y
        for i in range(1,len(rock_formations)):
            if rock_formations[i][0] == rock_formations[i-1][0]:
                stride = min(rock_formations[i-1][1], rock_formations[i][1]), max(rock_formations[i-1][1], rock_formations[i][1])
                for y in range(stride[0], stride[1]+1):
                    rockmap[rock_formations[i][0], y] = "#"
            else:
                stride = min(rock_formations[i-1][0], rock_formations[i][0]), max(rock_formations[i-1][0], rock_formations[i][0])
                for x in range(stride[0], stride[1]+1):
                    rockmap[x, rock_formations[i][1]] = "#"
    return (rockmap, (min_x,max_x,min_y,max_y))
            
                
                
                
```

```python
def is_blocked(pos: tuple , rockmap: dict):
    return pos in rockmap

def sand_step(pos, rockmap):
    if is_blocked((pos[0], pos[1]+1), rockmap):
        if is_blocked((pos[0]-1, pos[1]+1), rockmap):
            if is_blocked((pos[0]+1, pos[1]+1), rockmap):
                return pos
            else:
                return (pos[0]+1, pos[1]+1)
        else:
            return (pos[0]-1, pos[1]+1)
    else:
        return (pos[0], pos[1]+1)
```

```python
import copy

def drop_sand(rockmap, ymax):
    pos = (500, 0)
    while True:
        newpos = sand_step(pos, rockmap)
        if newpos == pos:
            rockmap[pos] = "o"
            return True
        if newpos[1] >= ymax:
            return False
        pos = newpos
        
def fill_up(rockmap, ymax):
    new_rockmap = copy.deepcopy(rockmap)
    sand_count = 0
    while drop_sand(new_rockmap, ymax): 
        sand_count += 1
    return (sand_count, new_rockmap)
        
```

```python
def rockmap_to_str(rockmap: dict, limits: tuple):
    min_x,max_x,min_y,max_y = limits
    for k in rockmap.keys():
        if k[0] < min_x: min_x = k[0]
        if k[0] > max_x: max_x = k[0]
    pic = ""
    for y in range(0, max_y):
        row = ""
        for x in range(min_x, max_x+1):
            row += rockmap.get((x,y), ".")
        pic += row + "\n"
    return pic
```

```python
def sand_step_p2(pos, rockmap, ymax):
    if pos[1] == ymax+1:
        return pos
    if is_blocked((pos[0], pos[1]+1), rockmap):
        if is_blocked((pos[0]-1, pos[1]+1), rockmap):
            if is_blocked((pos[0]+1, pos[1]+1), rockmap):
                return pos
            else:
                return (pos[0]+1, pos[1]+1)
        else:
            return (pos[0]-1, pos[1]+1)
    else:
        return (pos[0], pos[1]+1)

def drop_sand_p2(rockmap, ymax):
    pos = (500, 0)
    while True:
        newpos = sand_step_p2(pos, rockmap, ymax)
        if newpos == pos:
            rockmap[pos] = "o"
            return True
        pos = newpos
        
def fill_up_p2(rockmap, ymax):
    new_rockmap = copy.deepcopy(rockmap)
    sand_count = 0
    while not is_blocked((500,0),new_rockmap) and drop_sand_p2(new_rockmap, ymax): 
        sand_count += 1
    return (sand_count, new_rockmap)
```
